```python
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()

        self.conv1 = nn.Conv2d(3, 1000, 3, 1, 1, bias=False) # 200x200
        self.batchnorm1 = nn.BatchNorm2d(1000)
        self.pool2x2 = torch.nn.MaxPool2d(kernel_size=2, stride=2) # 100x100
        self.conv2 = nn.Conv2d(1000, 500, 3, 1, 1, bias=False)
        self.batchnorm2 = nn.BatchNorm2d(500)
        # Pool
        self.conv3 = nn.Conv2d(500, 100, 3, 1, 1, bias=False) # 50x50
        self.batchnorm3 = nn.BatchNorm2d(100)
        self.conv4 = nn.Conv2d(100, 80, 6, 1, 0, bias=False) # 45x45
        self.batchnorm4 = nn.BatchNorm2d(80)
        self.conv5 = nn.Conv2d(80, 60, 6, 1, 0, bias=False) # 40x40
        self.batchnorm5 = nn.BatchNorm2d(60)
        self.conv6 = nn.Conv2d(60, 50, 6, 1, 0, bias=False) # 35x35
        self.batchnorm6 = nn.BatchNorm2d(50)
        self.conv7 = nn.Conv2d(50, 40, 6, 1, 0, bias=False) # 30x30
        self.batchnorm7 = nn.BatchNorm2d(40)
        self.conv8 = nn.Conv2d(40, 30, 6, 1, 0, bias=False) # 25x25
        self.batchnorm8 = nn.BatchNorm2d(30)
        self.conv9 = nn.Conv2d(30, 20, 6, 1, 0, bias=False) # 20x20
        self.batchnorm9 = nn.BatchNorm2d(20)
        # Pool
        self.conv10 = nn.Conv2d(20, 10, 3, 1, 1, bias=False) # 10x10
        self.batchnorm10 = nn.BatchNorm2d(10)
        # Pool
        self.conv11 = nn.Conv2d(10, 4, 5, 1, 0, bias=False) # 1x1
        #self.batchnorm11 = nn.BatchNorm2d(8)
        #self.conv12 = nn.Conv2d(8, 4, 5, 1, 0, bias=False) # 1x1
        #self.conv13 = nn.Conv2d(25, 1, 1, 1, 0, bias=False)
        #self.sigmoid = nn.Sigmoid()
        
        self.Relu = nn.ReLU()
        self.LRelu = nn.LeakyReLU(0.2)
    
    def forward(self, input):
        x = self.conv1(input)
        x = self.batchnorm1(x)
        x = self.LRelu(x)
        
        x = self.pool2x2(x)
        
        x = self.conv2(x)
        x = self.batchnorm2(x)
        x = self.LRelu(x)
        
        x = self.pool2x2(x)
        
        x = self.conv3(x)
        x = self.batchnorm3(x)
        x = self.LRelu(x)
        x = self.conv4(x)
        x = self.batchnorm4(x)
        x = self.LRelu(x)
        x = self.conv5(x)
        x = self.batchnorm5(x)
        x = self.LRelu(x)
        x = self.conv6(x)
        x = self.batchnorm6(x)
        x = self.LRelu(x)
        x = self.conv7(x)
        x = self.batchnorm7(x)
        x = self.LRelu(x)
        x = self.conv8(x)
        x = self.batchnorm8(x)
        x = self.LRelu(x)
        x = self.conv9(x)
        x = self.batchnorm9(x)
        x = self.LRelu(x)
        
        x = self.pool2x2(x)
        
        x = self.conv10(x)
        x = self.batchnorm10(x)
        x = self.LRelu(x)
        
        x = self.pool2x2(x)
        
        x = self.conv11(x)
        #x = self.batchnorm11(x)
        #x = self.LRelu(x)
        
        #x = self.conv12(x)
        
        output = self.Relu(x)

        return output
```

```python
Emboxer = Discriminator().to(device)

optimizer = optim.Adam(Emboxer.parameters(), lr=1e-3, betas=(0, 0.99))
scheduler = optim.lr_scheduler.StepLR(optimizer, 50, gamma=0.1)

summary(Emboxer, (3, 200, 200))
```

```python
def train(dataset=None, epochs=1000, batch_size=6,loss=nn.MSELoss(), optimizer=optimizer, checkpoint=5000):
    
    dataloader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True)
    
    Emboxer.train(True)
    
    best_loss = float("inf")

    for epoch in range(epochs):
        
        epoch_loss = 0
        for item, (data, labels) in enumerate(dataloader):
            Emboxer.zero_grad()
            output = Emboxer(data.to(device).float())
            output = output.view(output.size(0), -1)
            #output = output.squeeze(-1)
            #output = output.squeeze(-1)
            #print(output.size())
            #print(output)
            labels = labels.to(device)
            #labels = labels.to(device)
            #labels = labels.float()
            #print(labels.size())
            #print(labels)

            cost = loss(output, labels)

            cost.backward()
            optimizer.step()
            
            epoch_loss += cost.item()
            
            grads_avg = torch.mean(Emboxer.conv11.weight.grad)

            if epoch_loss < best_loss:
                best_loss = cost
                best_params = Emboxer.state_dict()
    
            if item % checkpoint == 0 or epoch == epochs-1:
                print(f"{epoch}|{epochs}")
                print(f"Model Loss: {epoch_loss}\tBest Loss: {best_loss}\tGradients Average: {grads_avg}")
                print(f"Current LR: {scheduler.get_last_lr()[0]}")

                try:
                    torch.save(best_params, f'Face_Extractor.pth')
                
                except:
                    pass

                print("Model saved!")
                
    scheduler.step()
```
